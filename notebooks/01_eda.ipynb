# Telco Customer Churn - Exploratory Data Analysis

This notebook performs initial data exploration on the Telco Customer Churn dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

df = pd.read_csv("data/raw/WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

## Data Overview

```python
print(f"Shape: {df.shape}")
print(f"\nData types:\n{df.dtypes}")
print(f"\nMissing values:\n{df.isnull().sum()}")
print(f"\nClass balance:\n{df['Churn'].value_counts(normalize=True)}")
```

## Churn by Contract Type

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})

contract_churn = df.groupby('Contract')['Churn'].mean().sort_values(ascending=False)
print(contract_churn)

sns.barplot(x=contract_churn.index, y=contract_churn.values)
plt.title('Churn Rate by Contract Type')
plt.ylabel('Churn Rate')
plt.show()
```

## Churn by Tenure

```python
sns.histplot(data=df, x='tenure', hue='Churn', bins=30, kde=True)
plt.title('Tenure Distribution by Churn')
plt.xlabel('Tenure (months)')
plt.show()
```

## Churn by Monthly Charges

```python
sns.boxplot(data=df, x='Churn', y='MonthlyCharges')
plt.title('Monthly Charges by Churn')
plt.xticks([0, 1], ['No Churn', 'Churn'])
plt.show()
```

## Key Findings

- Month-to-month contracts have the highest churn rate (~45%).
- Customers with very short tenure churn significantly more.
- Higher monthly charges correlate with higher churn.
